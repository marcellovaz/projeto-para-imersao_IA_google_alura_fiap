```python
import datetime
import google.generativeai as genai
from google.colab import userdata

animais = []

api_key = userdat.get ('chave_secreta')

genai.configure(api_key= api_key)

def cadastrar_animal():
    """Coleta os dados do animal e adiciona à lista 'animais'."""

    especie = input("Espécie do animal: ")
    raca = input("Raça do animal: ")
    cor_pelo = input("Cor do pelo: ")
    sexo = input("Sexo (M/F): ").upper()
    caracteristica = input("Característica única (deixe em branco se não houver): ")
    placa_identificacao = input("Placa de identificação (deixe em branco se não houver): ")

    animal = {
        "especie": especie,
        "raca": raca,
        "cor_pelo": cor_pelo,
        "sexo": sexo,
        "caracteristica": caracteristica,
        "placa_identificacao": placa_identificacao,
        "proprietario": None
    }

    if placa_identificacao:
        nome_proprietario = input("Nome do Proprietário: ")
        telefone_proprietario = input("Telefone do Proprietário: ")
        animal["proprietario"] = {
            "nome": nome_proprietario,
            "telefone": telefone_proprietario
        }

    animais.append(animal)
    print("Animal cadastrado com sucesso!")

def consultar_por_cor_caracteristica():
    """Busca animais por cor do pelo e característica."""

    cor = input("Digite a cor do pelo: ")
    caracteristica = input("Digite a característica (deixe em branco se não houver): ")

    encontrados = []
    for animal in animais:
        if animal["cor_pelo"] == cor and (not caracteristica or animal["caracteristica"] == caracteristica):
            encontrados.append(animal)

    if encontrados:
        print("\nAnimais encontrados:")
        for animal in encontrados:
            print(f"- Espécie: {animal['especie']}, Raça: {animal['raca']}, Sexo: {animal['sexo']}")
    else:
        print("\nNenhum animal encontrado com essa cor e característica.")

def consultar_quantidade_por_raca():
    """Conta a quantidade de animais por raça."""

    contagem_racas = {}
    for animal in animais:
        raca = animal["raca"]
        if raca in contagem_racas:
            contagem_racas[raca] += 1
        else:
            contagem_racas[raca] = 1

    print("\nQuantidade de animais por raça:")
    for raca, quantidade in contagem_racas.items():
        print(f"- {raca}: {quantidade}")

def exibir_todos_animais():
    """Imprime uma tabela com todos os animais cadastrados."""

    print("\nLista de Animais Cadastrados:")
    print("-" * 50)
    print("{:<10} {:<15} {:<10} {:<10} {:<25}".format("Espécie", "Raça", "Cor", "Sexo", "Característica"))
    print("-" * 50)

    for animal in animais:
        print("{:<10} {:<15} {:<10} {:<10} {:<25}".format(
            animal["especie"],
            animal["raca"],
            animal["cor_pelo"],
            animal["sexo"],
            animal["caracteristica"]
        ))
    print("-" * 50)

while True:
    print("\nEscolha uma opção:")
    print("1 - Cadastrar Animal")
    print("2 - Consultar por Cor e Característica")
    print("3 - Consultar Quantidade por Raça")
    print("4 - Exibir Todos os Animais")
    print("0 - Sair")

    opcao = input("Opção: ")

    if opcao == "1":
        cadastrar_animal()
    elif opcao == "2":
        consultar_por_cor_caracteristica()
    elif opcao == "3":
        consultar_quantidade_por_raca()
    elif opcao == "4":
        exibir_todos_animais()
    elif opcao == "0":
        break
    else:
        print("Opção inválida!")
```
